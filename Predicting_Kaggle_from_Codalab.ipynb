```python
import pandas as pd
codalab_data = pd.read_csv('trial_data.csv')
humor_data = codalab_data.iloc[:, 1:10]
humor_data
```

```python
from sklearn.impute import SimpleImputer 
#print(humor_data.isnull().sum())
#constant_imputer=SimpleImputer(strategy='constant', fill_value=0)
constant_imputer=SimpleImputer(strategy='most_frequent')
humor_data.iloc[:,:]=constant_imputer.fit_transform(humor_data)
humor_data
#print(humor_data.isnull().sum())
```

```python
# logistics Regression
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
col = ['funniness_male']
x = humor_data[col]
#x = x.values.reshape(-1,1)

y = humor_data['is_humor_all']
#y = y.values.reshape(-1,1)
# split for 50:10 data
#X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.16,random_state=1111)
#len(X_train)

logreg = LogisticRegression(random_state=101)
scores_accuracy = cross_val_score(logreg, x,y, cv=5, scoring='accuracy')
scores_log_loss = cross_val_score(logreg, x, y, cv=5, scoring='neg_log_loss')
print("scores_accuracy Array: ", scores_accuracy)
print(logreg.__class__.__name__+" average accuracy is %2.3f" % scores_accuracy.mean())
print(logreg.__class__.__name__+" average log_loss is %2.3f" % -scores_log_loss.mean())
```

```python
# linear regression on is_humor as y
col = ['funniness_male']
x = humor_data[col]
new_col = ['is_humor_all']
y = humor_data[new_col]
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x, y)
r_sq = linreg.score(x, y)
print('coefficient of determination:', r_sq)
print('intercept:', linreg.intercept_)
print('slope:', linreg.coef_)
y_pred = linreg.predict(x)
print(y_pred)
```
